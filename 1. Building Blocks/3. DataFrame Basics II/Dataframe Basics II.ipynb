## Filtering dataframes with one condition

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.sex.head(10)
```

```python
titanic.sex == "male"  # panda series
```

```python
titanic[titanic.sex == "male"]["fare"]  # getting all rows with sex= 'male' and column fare, this is called chained indexing and not recommended
```

```python
titanic.loc[titanic.sex == "male", "fare"] # works same as above, this is recommended
```

```python
mask1 = titanic.sex == 'male'
mask1
```

```python
titanic_male = titanic.loc[mask1]
```

```python
titanic_male.head()
```

```python
titanic.dtypes
```

```python
mask2 = titanic.dtypes == object     #filering with dtype object
mask2
```

```python
titanic.loc[:, ~mask2]   # ~ sis opposite of mask2, so all columns with numerical values
```

```python
titanic.loc[mask1, ~mask2]  # combining both
```

## Fitering datadrames with multiple conditions (AND)

### Checking the hypothesis of whether being an adult male on the titanic was bad for your survival

### Conclusion: True

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
mask1 = titanic.sex == 'male'
mask1.head()
```

```python
mask2 = titanic.age > 14
mask2.head()
```

```python
(mask1 & mask2).head()
```

```python
male_surv = titanic.loc[mask1 & mask2, ['survived', 'pclass', 'sex', 'age']]
male_surv.head(20)
```

```python
male_surv.info()
```

```python
male_surv.describe()  #only 17% of adult males survived
```

```python
titanic.describe() # in total 38% survived
```

## Fitering datadrames with multiple conditions (AND)

### Checking whether you being a woman or a child helped in survivability

### Conclusion: True

```python
import pandas as pd
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
mask1 = titanic.sex == 'female'
```

```python
mask2 = titanic.age < 14
```

```python
wom_or_chil = titanic.loc[mask1 | mask2, ['survived', 'pclass', 'sex', 'age' ]]
```

```python
wom_or_chil.head()
```

```python
wom_or_chil.info()
```

```python
wom_or_chil.describe()  # 72% women or children survived
```

## Advance filtering with between(), isin() and ~

```python
import pandas as pd
```

```python
summer = pd.read_csv("summer.csv")
```

```python
summer.head()
```

```python
og_1988 =  summer.loc[summer.Year == 1988]  # filtering all data from 1988
```

```python
og_1988.head()   
```

```python
og_1988.tail()
```

```python
og_1988.info()
```

```python
og_since1992 = summer.loc[summer.Year >= 1992]
```

```python
og_since1992.head()
```

```python
og_since1992.tail()
```

```python
summer.Year.between(1960, 1969).head()    
```

```python
og_60s = summer.loc[summer.Year.between(1960, 1969)]   # both 1960 and 1969 are included if inclusive parameter = True
```

```python
og_60s.head()
```

```python
og_60s.tail()
```

```python
my_fav_games = [1972, 1996]
```

#### isin() method

```python
summer.Year.isin(my_fav_games).head()
```

```python
og_72_96 = summer.loc[summer.Year.isin(my_fav_games)]
```

```python
og_72_96.head()
```

```python
og_72_96.tail()
```

```python
og_not_72_96 = summer.loc[~summer.Year.isin(my_fav_games)]
```

```python
og_not_72_96.Year.unique()
```

## any() an all()

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.sex == 'male'
```

```python
(titanic.sex == 'male').any()  # checking if any of the elements in the series is not zero(false)
```

```python
(titanic.sex == 'male').all()  #checking if all of the elements in the series is not zero(false)
```

```python
(titanic.sex == '80').any()  # checking if any of the passengers is 80 yo
```

```python
pd.Series([-1, 0.5, 1, -0.1, 0]).all()   # can also be used for numerical values true = 1
```

```python
titanic.fare.all()  # checking if all passengers paid something for the ticket
```
