# Visualisation with Matplotlib

### the plot() method

```python
import pandas as pd
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot(subplots = True, figsize = (15, 12), sharex = False, sharey = False)
plt.show()
```

```python
titanic.age.plot(figsize = (12, 8))
plt.show()
```

## Customisation 

```python
titanic.age.plot(figsize = (12,8), fontsize = 13, c = 'red', linestyle = '-', xlim = (0,900), ylim = (0,80), xticks = xticks, yticks = yticks , rot = 45)
plt.title('Titanic- Ages', fontsize = 15)
plt.legend(loc = 'best', fontsize = 15)
plt.xlabel('Passenger No', fontsize = 13)
plt.ylabel('Age', fontsize = 13)
#plt.grid()
plt.show()
```

```python
plt.style.available    # different styles of plots
```

```python
plt.style.use('classic')
```

```python
xticks = [x for x in range(0,901, 50)]    #defining the ticks
```

```python
yticks = [y for y in range(0,81, 5)]
```

## Histograms

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.head()
```

```python
titanic.age.value_counts(bins = 10)
```

```python
titanic.age.plot(kind = 'hist', figsize = (12,8), fontsize = 15, bins = 80)
plt.show()
```

using direct hist( ) method:

```python
titanic.age.hist(figsize = (12,8), bins = 80, xlabelsize = 15, ylabelsize = 15)
plt.show()
```

another way:

```python
plt.figure(figsize = (12,8))
plt.hist(titanic.age.dropna(), bins = 80, density = True, cumulative = True)   # if density =true, it's relative quantities
plt.show()
```

## Barcharts and Piecharts

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
summer_2012 = pd.read_csv('summer_2012.csv', index_col = 'Country')
```

```python
summer_2012
```

```python
summer_2012.Medal.plot(kind = 'barh', figsize =(12,8), fontsize = 12, ylabel = 'Medals')  # barh for horizontal bar chart
plt.show()
```

```python
summer_2012.Medal.plot(kind = 'pie', figsize =(12,8), fontsize = 12) 
plt.show()
```
