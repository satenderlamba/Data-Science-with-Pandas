## Merging, joining and concatenating dataframes

### Adding/ Concatenating rows part 1

```python
import pandas as pd
```

```python
men2008 = pd.read_csv('men2004.csv')
```

```python
men2008
```

```python
men2004 = pd.read_csv('men2004.csv')
```

```python
men2004
```

first method (less powerful)

```python
men2004.append(men2008, ignore_index = True) # append another dataframe
```

second more powerful method pd.concat():

```python
pd.concat([men2004, men2008], ignore_index = False, keys = [2004, 2008], names = ['Year'])   # combine horizontally if axis = 1
```

```python
men0408 = pd.concat([men2004, men2008], ignore_index = False, keys = [2004, 2008], names = ['Year'])
```

```python
men0408
```

```python
men0408.reset_index()
```

```python
men0408.reset_index().drop(columns = 'level_1')  # dropping column level 1
```

```python
men2004.drop(labels = ['Success'], axis = 1, inplace = True)
```

```python
men2004   # sucsess column dropped
```

let's match the column names as well

```python
men2008.columns = men2004.columns
```

```python
men2008.head()    # same column labels as men2004
```

```python
pd.concat([men2004, men2008], keys = [2004, 2008], names = ['Year'])         # nice and clean now!
```

## Arithmetic b/w Pandas Objects / Data Alignment

```python
import pandas as pd
```

```python
topfive_2004 = pd.read_csv('topfive_2004.csv', index_col = 'Athlete')
topfive_2008 = pd.read_csv('topfive_2008.csv', index_col = 'Athlete')
```

```python
topfive_2004
```

```python
topfive_2008
```

2004 has 'Bronze' but 2008 has 'bronze' column, problem!!

simple arithmetic '+', not recommended

```python
topfive_2004 + topfive_2008          # simple arithmetic, many NaN values cuz there are diff column labels
```

better method to add:

```python
topfive_2004.add(topfive_2008, fill_value = 0)     # fill_value fills the value in missing values
```

^^ there are still missing values as the column labels are not same

let's fix the name 'bronze' in topfive_2008

```python
topfive_2008.rename(columns = {'bronze': 'Bronze'}, inplace = True)
```

```python
topfive_2004.add(topfive_2008, fill_value = 0)
```

## ^.^ Nice

there are also 'sub', 'mul', and 'div' methods for subtraction, multiplication and division respectiely.

## Outer Joins with merge( )

```python
import pandas as pd
```

```python
men2004 = pd.read_csv('men2004.csv')     # swimming event medal winners of 2004 & 2008
men2008 = pd.read_csv('men2008.csv')
```

```python
men2004.head()
```

```python
men2004.shape
```

```python
men2008.head()
```

```python
men2008.shape
```

Outer Join returns all athletes who were Successful in at least one Edition.

```python
len(men2004) + len(men2008)
```

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete')   # on gets the column(s) on which to apply outer join on            
```

Let's fix the column labels using 'suffixes' param, and indicator param

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

merge column is very useful in knowing which atletes were successful in which year(s)

```python
men0408 = men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

```python
men0408._merge.value_counts()
```

^^ Very interesting and useful info.
