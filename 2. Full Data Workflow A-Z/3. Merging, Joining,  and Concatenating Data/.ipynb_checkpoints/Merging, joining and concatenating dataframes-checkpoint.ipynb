## Outer Joins with merge( )

```python
import pandas as pd
```

```python
men2004 = pd.read_csv('men2004.csv')     # swimming event medal winners of 2004 & 2008
men2008 = pd.read_csv('men2008.csv')
```

```python
men2004.head()
```

```python
men2004.shape
```

```python
men2008.head()
```

```python
men2008.shape
```

Outer Join returns all athletes who were Successful in at least one Edition.

```python
len(men2004) + len(men2008)
```

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete')   # on gets the column(s) on which to apply outer join on            
```

Let's fix the column labels using 'suffixes' param, and indicator param

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

merge column is very useful in knowing which atletes were successful in which year(s)

```python
men0408 = men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

```python
men0408._merge.value_counts()
```

^^ Very interesting and useful info.

## Inner Join

```python
men2004.head()
```

```python
men2008.head()
```

Inner join returns the Athletes succesful in both years.

```python
men2004.merge(men2008, how = 'inner', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

^^ Nice

## Outer join without Intersection

```python
men2004.head()
```

```python
men2008.head()
```

outer join returns all athletes who were successful in only one year.

```python
men0408 = men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)
```

```python
men0408
```

we only need to filter this to exclude 'both' in _merge column

```python
men0408.loc[men0408._merge != 'both']         # this is the req dataframe
```

## Left join without intersection

```python
men2004.head()
```

```python
men2008.head()
```

## Right Join without Intersection

```python
men0408.loc[men0408._merge == 'right_only']    # required dataset
```

## Left Join    (most important and most used)

```python
men2004.head()
```

```python
men2008.head()
```

Left join returns all athletes who were succ. in 2004, and include 2008 medals as well.

we use the 'left' in how param

```python
men2004.merge(men2008, how = 'left', on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)   # required dataframe
```

## Right Join

Right join returns all athletes who were succ. in 2008, and include 2004 medals as well.

```python
men2004.merge(men2008, how = 'right', on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)   # required dataframe
```

^^ you can also use left join to get this result but change the dataframe order first.

```python
men2008.merge(men2004, how = 'left', on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)   # same result as above
```

## Joining on different Column Labels and Indexes

```python
import pandas as pd
```

```python
men2004 = pd.read_csv('men2004.csv')
men2008 = pd.read_csv('men2008.csv')
```

```python
men2004.columns = ['Name', 'Medals']   # changing the column names so they're different.
```

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ['_2004, _2008'], indicator = True)   # error as Athlete is not in both dfs
```

let's fix this

```python
men0408 = men2004.merge(men2008, how = 'outer', left_on = 'Name', right_on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)
```

```python
men0408    # done, 2 seperate columns Name and Athlete
```

```python
men0408.Name.fillna(men0408.Athlete, inplace = True)    # filling missing values in Name
```

```python
men0408
```

```python
men0408.drop(['Athlete', '_merge'], axis = 1, inplace = True)
```

```python
men0408    # Clean!!
```

```python
men2004.head()
```

```python
men2008.set_index('Athlete', inplace = True)   # setting athlete as index
```

```python
men2008
```

```python
men2004.merge(men2008, how = 'outer', left_on = 'Name', right_on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)
```

 ^^lost index labels in right dataframe !??!?!

let's use the right index as join key:

```python
men2004.merge(men2008, how = 'outer', left_on = 'Name', right_index = True, suffixes = ['_2004', '_2008'], indicator = True)
```

^^ Noice!

# Joining on many columns (real world case)

```python
import pandas as pd
```

```python
men2004_det = pd.read_csv('men2004_det.csv')    # det = detailed
men2008_det = pd.read_csv('men2008_det.csv')
```

```python
men2004_det.head(10)   # same dataframe but detailed
```

```python
men2008_det.head(10)
```

```python
men2004_det.loc[men2004_det.Athlete == 'PHELPS, Michael']
```

```python
men2008_det.loc[men2008_det.Athlete == 'PHELPS, Michael']
```

let's merge the two:

```python
men0408 = men2004_det.merge(men2008_det, how = 'outer', on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)
```

```python
men0408.loc[men0408.Athlete == 'PHELPS, Michael']    # weird, not what we want
```

let's fix this by passing 2 columns to 'on' param

```python
men0408 = men2004_det.merge(men2008_det, how = 'outer', on = ['Athlete', 'Medal'], suffixes = ['_2004', '_2008'], indicator = True)
```

```python
men0408.loc[men0408.Athlete == 'PHELPS, Michael']
```

^^ NIce!

now let's try inner join! (Intersection)

```python
men0408 = men2004_det.merge(men2008_det, how = 'inner', on = ['Athlete', 'Medal'], suffixes = ['_2004', '_2008'], indicator = True)
```

```python
men0408.loc[men0408.Athlete == 'PHELPS, Michael']
```

^^as expected

## pd.merge( ) and join( )

```python
import pandas as pd
```

```python
men2004.head()
```

```python
men2008.head()
```

Outer join:

```python
men2004.merge(men2008, how = 'outer', on = 'Athlete', suffixes = ['_2004', '_2008'], indicator = True)
```

pd.merge()

```python
pd.merge(men2004, men2008, how = 'outer', on = 'Athlete', suffixes = ('_2004', '_2008'), indicator = True)  # same as above
```

join() method, works best only when applying on dfs with same index names, not much functionality. Not recommended.
